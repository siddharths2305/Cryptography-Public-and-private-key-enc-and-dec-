```python
import random

def generate_prime_number():
    # Generate a large prime number
    while True:
        p = random.randint(2**10, 2**11)
        if is_prime(p):
            return p

def is_prime(n):
    # Check if a number is prime
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def calculate_primitive_root(p):
    # Find a primitive root modulo p
    primitive_roots = []
    for g in range(2, p):
        if is_primitive_root(g, p):
            primitive_roots.append(g)
    return random.choice(primitive_roots)

def is_primitive_root(g, p):
    # Check if g is a primitive root modulo p
    values = set()
    for i in range(1, p):
        values.add(pow(g, i, p))
    return len(values) == p - 1

def generate_private_key(p):
    # Generate a private key
    return random.randint(2, p - 2)

def generate_public_key(g, private_key, p):
    # Generate a public key
    return pow(g, private_key, p)

def generate_shared_secret(public_key, private_key, p):
    # Generate a shared secret
    return pow(public_key, private_key, p)

# Generate prime number and primitive root
p = generate_prime_number()
g = calculate_primitive_root(p)

# Generate private and public keys for Alice
private_key_alice = generate_private_key(p)
public_key_alice = generate_public_key(g, private_key_alice, p)

# Generate private and public keys for Bob
private_key_bob = generate_private_key(p)
public_key_bob = generate_public_key(g, private_key_bob, p)

# Exchange public keys
shared_secret_alice = generate_shared_secret(public_key_bob, private_key_alice, p)
shared_secret_bob = generate_shared_secret(public_key_alice, private_key_bob, p)

# Check if the shared secrets match
print("Shared secret (Alice):", shared_secret_alice)
print("Shared secret (Bob):", shared_secret_bob)


```

```python
def encrypt():
    password=code.get()
    
    if password==keyE:       
        message=text1.get(1.0,END)
        encode_message=message.encode("ascii")
        base64_bytes=base64.b64encode(encode_message)
        encrypt=base64_bytes.decode("ascii")
        
        f1.insert(END,encrypt,'encrt')
        f1.insert(END,"\n....................Encrypted successfully .../\n",'encrt')
        
    elif password=="":
        messagebox.showerror("encryption","Input Password")
        f1.insert(END,"\n _encryption key is unavailable (empty).../\n")
    elif password !=keyE:
        messagebox.showerror("encryption","INVALID Password")
        messagebox.showerror("encryption","Encryption Denied")
        f1.insert(END,"\n _encryption key is incorrect (WARNING).../\n")
        
def decrypt():
    password=code.get()
    if password==keyD:
        message=text1.get(1.0,END)
        decode_message=message.encode("ascii")
        base64_bytes=base64.b64decode(decode_message)
        decrypt=base64_bytes.decode("ascii")
        f1.insert(END,decrypt,'decrp')
        f1.insert(END,"\n....................Decrypted successfully .../\n",'decrp')
        
    elif password=="":
        messagebox.showerror("decryption","Input Password")
        f1.insert(END,"\n _decryption key is unavailable (empty).../\n")    
    
    elif password !=keyD:
        messagebox.showerror("decryption","INVALID Password")
        messagebox.showerror("decryption","Decryption Denied")
        f1.insert(END,"\n _decryption key is incorrect (WARNING).../\n")
        
```
